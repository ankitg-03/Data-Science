### Importing Required Modules

# Analysis

```python
x = data.groupby("Year").sum()
x
```

## Time-Series Analysis of Crimes During the Year 2001 - 2014

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_size_inches(13, 5)

x["Rape"].plot(color = "red", ax = ax1)
x["Kidnapping and Abduction"].plot(color = "red", ax = ax2)

ax1.set_title("Rape Cases")
ax1.grid()

ax2.set_title("Kidnapping and Abduction Cases")
ax2.grid()

ax1.set(ylabel = "Cases")
ax2.set(ylabel = "Cases")

plt.tight_layout()
plt.show()
```

***Rape***
* Number of cases have increased over time. 
* There is a dip in number of cases in year 2002-2003.
* Highest jump can be seen in year 2012-2013.

 ***Kidnapping/Abduction***
 * There is a  constant rise in the number of cases.
 * A significant rise in the number of cases have been noticed in the years 2012-2014.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_size_inches(13, 5)

x["Dowry Deaths"].plot(color = "red", ax = ax1)
x["Assault on women with intent to outrage her modesty"].plot(color = "red", ax = ax2)

ax1.set_title("Dowry Deaths")
ax1.grid()

ax2.set_title("Assault on women with intent to outrage her modesty")
ax2.grid()

ax1.set(ylabel = "Cases")
ax2.set(ylabel = "Cases")

plt.tight_layout()
plt.show()
```

***Dowry Deaths***
* Number of Dowry Deaths cases have decreased in years 2001-2003,2004-2005 & 2011-2013.
* There has been a significant rise from the year 2005-2011 and 2013-2014.

***Assault on women with intent to outrage her modesty***
* A slight decrease in number of cases can be seen in the years 2001-2003 and 2008-2009
* A steep rise is seen from the year 2012-2014

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

fig.set_size_inches(13, 5)

x["Importation of Girls"].plot(color = "red", ax = ax1)
x["Cruelty by Husband or his Relatives"].plot(color = "red", ax = ax2)

ax1.set_title("Importation of Girls")
ax1.grid()

ax2.set_title("Cruelty by Husband or his Relatives")
ax2.grid()

ax1.set(ylabel = "Cases")
ax2.set(ylabel = "Cases")

plt.tight_layout()
plt.show()
```

***Importation of Girls***
* There has been a noticeable rise in number of cases during the years 2003-2005 and 2010-2011
* Overall number of cases has been reduced over time.

***Cruelty by Husband or his Relatives***
* Crimes under 'Cruelty by Husband or his Relatives' has continuously increased over time

```python
plt.figure(figsize = (12,5))
x["Insult to modesty of Women"].plot(color = "red")
plt.title("Insult to modesty")
plt.ylabel("Cases")
plt.grid()
plt.show()
```

***Insult to Modesty***
* Number of cases of Insult to modesty of women is fluctuating throughout the years.

## Total Crime Rate Distribution

```python
pi_list = [sum(x["Rape"]),sum(x["Kidnapping and Abduction"]),
           sum(x["Dowry Deaths"]),sum(x["Assault on women with intent to outrage her modesty"]),
           sum(x["Insult to modesty of Women"]),sum(x["Cruelty by Husband or his Relatives"]),sum(x["Importation of Girls"]) ]
```

```python
crimes_list = ['Rape', 'Kidnapping and Abduction','Dowry Deaths', 'Assault on women with intent to outrage her modesty',
       'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']

color_l = ["crimson", "steelblue", "darkorange", "thistle", "gold", "limegreen", "tan"]

plt.pie(pi_list, autopct = "%.2f%%", radius = 1.4, labels=crimes_list,
       colors = color_l)
plt.show()
```

* Cruelty by Husband or his Relatives accounts for nearly half(42%) of all the crimes against women.
* Importation of Girls has the least rate i.e. 0.04% among all other crimes against women throughout all the years.

## Insights

```python
inc_r = ((x.iloc[13,0] - x.iloc[0,0]) / x.iloc[0,0]) * 100 
inc_ka = ((x.iloc[13,1] - x.iloc[0,1]) / x.iloc[0,1]) * 100 
inc_dd = ((x.iloc[13,2] - x.iloc[0,2]) / x.iloc[0,2]) * 100 
inc_am = ((x.iloc[13,3] - x.iloc[0,3]) / x.iloc[0,3]) * 100 
inc_i = ((x.iloc[13,4] - x.iloc[0,4]) / x.iloc[0,4]) * 100 
inc_c = ((x.iloc[13,5] - x.iloc[0,5]) / x.iloc[0,5]) * 100 
inc_img = ((x.iloc[13,6] - x.iloc[0,6]) / x.iloc[0,6]) * 100
```

```python
inc_df = pd.DataFrame( {
                        "Cases" : crimes_list,
                        "Increase/Decrease Percentage(%) " : [inc_r,inc_ka,inc_dd,inc_am,inc_i,inc_c,inc_img]
                       } )
```

```python
def color(val): 

    if val > 0: 
        color = 'red'
    else: 
        color = 'green'
    return 'color: %s' % color 
  
inc_df.style.applymap(color,subset=['Increase/Decrease Percentage(%) '])
```

* 'Red' color indicates increase in crime rate.
*  Decrease in the crime rate has been highlighted in 'Green'.

*Kidnapping and Abduction cases recorded the maximum increase rate(~292%) among all other crimes.*

*Crime Rate of importation of Girls have decreased by approx. 89% over time.*


## Analysing Top 10 States with Highest Crime Rate

```python
new = data.groupby("STATE/UT").sum()
new.drop("Year", inplace = True, axis = 1)
new = new.T.sum()

plt.figure(figsize = (12,5))

new.sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'red')

plt.ylabel("Cases")
plt.title("Top 10 States with Maximum Cases")
plt.show()
```

*Uttar Pradesh  holds the worst record by laying claim to nearly 300000 cases of crime against women among top 10 states with highest crime rate in the country followed by Andhra Pradesh, West Bengal, Rajasthan & Madhya Pradesh as top 5 states in terms of worst women safety*

### Time-Series Analysis

```python
Z = data.groupby(["Year","STATE/UT"]).sum()
Z = Z.T
Z.head(3)
```

```python
Y = pd.DataFrame(Z.sum())
Y.reset_index(inplace = True)
Y.columns = ["Year", "State", "Cases"]
Y.head(3)
```

```python
plt.figure(figsize = (16,8))
sns.lineplot(x = "Year", y = "Cases", data = Y[Y["State"] == "Uttar Pradesh"], label = "Uttar Pradesh")
sns.lineplot(x = "Year", y = "Cases", data = Y[Y["State"] == "Andhra Pradesh"], label = "Andhra Pradesh")
sns.lineplot(x = "Year", y = "Cases", data = Y[Y["State"] == "West Bengal"], label = "West Bengal")
sns.lineplot(x = "Year", y = "Cases", data = Y[Y["State"] == "Rajasthan"], label = "Rajasthan")
sns.lineplot(x = "Year", y = "Cases", data = Y[Y["State"] == "Madhya Pradesh"], label = "Madhya Pradesh")
plt.title("Time-Series Analysis of Top 5 States with Maximum Cases")
plt.legend()
plt.grid()
plt.show()
```

* Crime rate has been increased over time in all the states.
* Significant dip of crimes against women is noticeable in Andhra Pradesh from year 2013-2014

## Analysing Top 20 Districts with Crimes against Women

```python
dist_df = data.groupby(["STATE/UT","DISTRICT"]).sum()
dist_df = pd.DataFrame(dist_df.T.sum())
dist_df.reset_index(inplace = True)
dist_df.columns = ["States","District","Cases"]
```

```python
new_df = dist_df.sort_values(by="Cases", ascending= False).reset_index(drop = True).head(20)
new_df
```

```python
plt.figure(figsize = (14,4))
sns.barplot(x = "District", y = "Cases", data = new_df)
plt.xticks(rotation = 90)
plt.title("Top 20 Districts with Maximum Cases")
plt.show()
```

*Murshidabad, district in West Bengal, has reported the maximum number of cases against women among all the districts*
