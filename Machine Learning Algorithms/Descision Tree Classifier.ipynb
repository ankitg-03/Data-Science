**importing required libraries**

```python
import pandas as pd

from sklearn import datasets
from sklearn import tree

import matplotlib.pyplot as plt
```

**loading the dataset**

```python
iris = datasets.load_iris()
```

```python
X = iris.data
Y = iris.target
```

```python
iris.DESCR
```

**Creating Decision Tree Classifier model**

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X,Y)
print("Decision Tree Classifier Created...")
```

**Predicting the output**

```python
model.predict([[1,2,3,4]])
```

**Plotting the decission tree**

```python
f_n = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'] #feature names
c_n = ['setosa', 'versicolor', 'virginica']  # class names

fig, ax = plt.subplots(figsize = (10,8), dpi = 300)
tree.plot_tree(model, feature_names=f_n, class_names=c_n, filled = True)
plt.show()
```
